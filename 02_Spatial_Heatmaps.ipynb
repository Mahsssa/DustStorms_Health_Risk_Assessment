## Introduction

Using this notebook, you can see a heatmap of elderly people who visited the hospital within 7 days after a dust event with respiratory and cardiac diseases in Ahwaz

```python
import numpy as np
import pandas as pd
import folium
from folium.plugins import HeatMap
from folium.plugins import FastMarkerCluster
import warnings 
warnings.filterwarnings('ignore')
```

```python
# importing datasets

path = r'..\Data\3_all_patients_spatial_dataset.csv'
df = pd.read_csv(path)

resp_path = R'..\Data\4_respiratory_spatial_dataset.csv'
resp_df = pd.read_csv(resp_path)

cardio_path = R'..\Data\5_cardiovascular_spatial_dataset.csv'
cardio_df = pd.read_csv(cardio_path)
```

```python
# definfing a function for generating base map

def ahvaz_basemap(default_location=[31.31, 48.68], default_zoom_start=11):
    basemap = folium.Map(location=default_location, zoom_start=default_zoom_start)
    return basemap
```

## Respirotery diseases Heat Maps

```python
resp_df_weights = pd.DataFrame()
resp_df_weights = resp_df['id'].value_counts().reset_index()
resp_df_weights.columns = ['id', 'weight']
```

```python
resp_df_locations = resp_df[['id','age_range','lon','lat']]
resp_df_locations.drop_duplicates(subset=['id'], keep='first', inplace=True)
```

```python
print(resp_df_weights.shape)
print(resp_df_locations.shape)
```

```python
resp_df_heatmap = resp_df_weights.merge(resp_df_locations, on='id', how='left')
#resp_df_heatmap.sample(10)
```

### >65 Years Old Respiratory Patients 

```python
resp_65 = resp_df_heatmap.loc[(resp_df_heatmap['age_range'] == '65-74') | (resp_df_heatmap['age_range'] == '>75')]
resp_65.shape
```

```python
basemap_5 = ahvaz_basemap()
HeatMap(resp_65[['lat', 'lon', 'weight']], overlay=False, radius=15, min_opacity=0.45).add_to(basemap_5)
basemap_5
```

```python
FastMarkerCluster(resp_65[['lat', 'lon', 'weight']], zoom=20).add_to(basemap_5)
basemap_5
```
